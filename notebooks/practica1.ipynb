```julia
using PlotlyJS
using LinearAlgebra
using Printf
```

```julia
a,b = 0,1
t_min, t_max = 0,0.25
ci = x -> 1-x
h = 0.1
k = 0.0005
λ = k/h^2
xn = Int((b-a)/h)
tn  = Int((t_max-t_min)/k)

di = fill(λ,xn)
d  = fill(1-2*λ,xn+1)
ds = fill(λ,xn)
A  = Tridiagonal(di, d, ds)
A[1,1] = 1 - λ
A[xn+1,xn+1] = 1 - λ -λ*h

S = zeros(tn+1,xn+1)
S[1,1:xn+1] = ci.(LinRange(a, b, xn+1))

for t in 2:tn+1
    S[t,1:xn+1] = A*S[t-1,1:xn+1] + k*exp(-t)*ones(xn+1)
end

println("x \t\t S(x,t) ")
for (index, value) in enumerate(S[tn+1,:])
    @printf("x_%d = %.2f \t S(x_%d,%.2f) = %.8f\n", index-1, a + (index-1)*h, index-1, t_max, value)
end
```
